1) Faça um código em Python para o seguinte problema

Pergunte quantas pessoas pagarão a excursão com desconto de 12,5%

Pergunte quantas pessoas pagarão a excursão com desconto de 25%

Pergunte quantas pessoas pagarão a excursão sem desconto

Determine o total arrecadado pela empresa sabendo que o valor da excursão sem desconto é de R$ 1250,00.

Faça o teste com os seguintes valores

Desconto de 12,5% = 20          Desconto de 25% = 32          Sem desconto = 42

Apresente a reposta da seguinte maneira:

O Valor arrecadado pela empresa de excursões foi R$ {resposta}. A resposta deve estar com 2 casas decimais


```python
# Entradas dos usuários
desconto_12_5 = int(input("Quantas pessoas pagarão a excursão com desconto de 12,5%? "))
desconto_25 = int(input("Quantas pessoas pagarão a excursão com desconto de 25%? "))
sem_desconto = int(input("Quantas pessoas pagarão a excursão sem desconto? "))

# Valor da excursão
valor_excursao = 1250.00

# Cálculo dos valores com desconto
valor_desconto_12_5 = valor_excursao * 0.125  # Valor do desconto de 12,5%
valor_desconto_25 = valor_excursao * 0.25    # Valor do desconto de 25%

# Cálculo do total arrecadado
total_arrecadado = (desconto_12_5 * (valor_excursao - valor_desconto_12_5)) + \
                   (desconto_25 * (valor_excursao - valor_desconto_25)) + \
                   (sem_desconto * valor_excursao)

# Exibição do valor arrecadado
print(f"O valor arrecadado pela empresa de excursões foi R$ {total_arrecadado:.2f}.")
```

3) Os dados a seguir correspondem aos pesos, em kg, dos sacos de um determinado produto. Faça um gráfico BoxPlot e determine:

a) se existe outliers.

b) Q1 e Q3

c) O intervalo interquartil

d) A mediana

1,01 0,85 1,11 1,12 1,14 1,14 1,15 1,15 1,10 1,18 1,12 1,15 0,79 1,01 1,03 1,05 1,08 0,62 1,08 1,09 1,01 1,20 1,21 1,05 1,08 1,09 1,10 1,15 1,12 1,15


```python
import numpy as np
import matplotlib.pyplot as plt


dados = [1.01, 0.85, 1.11, 1.12, 1.14, 1.14, 1.15, 1.15, 1.10, 1.18, 1.12, 1.15,
         0.79, 1.01, 1.03, 1.05, 1.08, 0.62, 1.08, 1.09, 1.01, 1.20, 1.21, 1.05,
         1.08, 1.09, 1.10, 1.15, 1.12, 1.15]


dados_sorted = np.sort(dados)


Q1 = np.percentile(dados_sorted, 25)
Q3 = np.percentile(dados_sorted, 75)


mediana = np.median(dados_sorted)

IQR = Q3 - Q1

limite_inferior = Q1 - 1.5 * IQR
limite_superior = Q3 + 1.5 * IQR


outliers = [x for x in dados_sorted if x < limite_inferior or x > limite_superior]

print(f"Q1 (1º Quartil): {Q1}")
print(f"Q3 (3º Quartil): {Q3}")
print(f"Mediana: {mediana}")
print(f"Intervalo Interquartil (IQR): {IQR}")
print(f"Outliers: {outliers}")


plt.boxplot(dados_sorted, vert=False)
plt.title('Gráfico BoxPlot dos Pesos')
plt.xlabel('Pesos (kg)')
plt.show()
```

4) Uma turma com 22 alunos realizou uma prova de física e obtiveram os seguintes resultados

7,8 - 5,0 - 3,1 - 4,8 - 4,4 - 5,4 - 8,5 - 5,9 - 5,9 - 3,1 - 5,6 - 3,7 - 6,6 - 8,4 - 4,3 - 5,3 - 3,2 - 7,6 - 4,2 - 8,3 - 2,3 - 6,4

Determine:

a) Calcule a média e o desvio padrão das notas.

b) Dado o intervalo [media - DP, media + DP] determine quantos alunos ficaram dentro desse intervalo, quantos alunos ficaram acima desse intervalo e quantos alunos ficaram abaixo desse intervalo. Use o comando print para expressar as respostas de forma clara.


```python
import numpy as np

notas = [7.8, 5.0, 3.1, 4.8, 4.4, 5.4, 8.5, 5.9, 5.9, 3.1, 5.6, 3.7, 6.6, 8.4,
         4.3, 5.3, 3.2, 7.6, 4.2, 8.3, 2.3, 6.4]


N = len(notas)


media = np.mean(notas)
desvio_padrao = np.std(notas)


intervalo_inferior = media - desvio_padrao
intervalo_superior = media + desvio_padrao


dentro_intervalo = sum(1 for nota in notas if intervalo_inferior <= nota <= intervalo_superior)
acima_intervalo = sum(1 for nota in notas if nota > intervalo_superior)
abaixo_intervalo = sum(1 for nota in notas if nota < intervalo_inferior)

print(f"Média das notas: {media:.2f}")
print(f"Desvio padrão das notas: {desvio_padrao:.2f}")
print(f"Intervalo [média - DP, média + DP]: [{intervalo_inferior:.2f}, {intervalo_superior:.2f}]")
print(f"Alunos dentro do intervalo: {dentro_intervalo}")
print(f"Alunos acima do intervalo: {acima_intervalo}")
print(f"Alunos abaixo do intervalo: {abaixo_intervalo}")
```
